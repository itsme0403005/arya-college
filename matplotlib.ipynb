## Scatter

```python
xpoint=np.array([1,3,5,7,9,2,4,6,7,9])
ypoint=np.array([1,3,5,7,9,8,6,4,2,9])
plt.scatter(xpoint,ypoint)
plt.show()
```

```python
#compare plot
x1=np.array([1,3,5,7,9,2,4,6,7,9,3])
y1=np.array([1,3,5,7,9,8,6,4,2,9,4])
plt.scatter(x1,y1)
x2=np.array([1,4,8,2,7,0,3,5,9,0,3])
y2=np.array([8,6,2,6,3,8,1,9,3,8,2])
plt.scatter(x2,y2)
```

```python
# change clr of particular data points
x1=np.array([1,3,5,7,3,2,6,1,7,1])
y1=np.array([4,2,9,4,5,2,7,1,6,8])
c=np.array(['red','green','blue','yellow','purple','pink','orange','violet','indigo','black'])
plt.scatter(x1,y1,c=c)
```

```python
#increase size of markers
x1=np.array([1,3,5,7,3,2,6,1,7,1])
y1=np.array([4,2,9,4,5,2,7,1,6,8])
c=np.array(['red','green','blue','yellow','purple','pink','orange','violet','indigo','black'])
size=np.array([10,20,30,40,50,60,70,80,90,10])
plt.scatter(x1,y1,c=c,s=size)
```

```python
# used alpha parameter to solve overlap 
x1=np.array([1,1,5,7,3,2,6,1,7,1])
y1=np.array([4,1,9,4,5,2,7,1,6,8])
c=np.array(['red','green','blue','yellow','purple','pink','orange','violet','indigo','black'])
size=np.array([10,200,30,40,50,60,70,80,90,10])
plt.scatter(x1,y1,c=c,s=size,alpha=0.5) # visibility
```

```python
x1=np.array([1,3,5,7,3,2,6,1,7,1])
y1=np.array([4,1,9,4,5,2,7,1,6,8])
c=np.array(['red','green','blue','yellow','purple','pink','orange','violet','indigo','black'])
size=np.array([10,200,30,40,50,60,70,80,90,10])
plt.scatter(x1,y1,c=c,s=size,alpha=0.5,cmap=' nipy_spectral') #used to draw colorbar
#cam=virdis,nipy_spectral
plt.colorbar()
```

```python
x1=np.array([1,3,5,7,3,2,6,1,7,1])
y1=np.array([4,1,9,4,5,2,7,1,6,8])
c=np.array(['red','green','blue','yellow','purple','pink','orange','violet','indigo','black'])
size=np.array([10,200,30,40,50,60,70,80,90,10])
plt.scatter(x1,y1,c=c,s=size,alpha=0.5,cmap='virdis') #used to draw colorbar
#cam=virdis,nipy_spectral
plt.colorbar()
```

## Barplot

```python
#vertical
x1=np.array(['a','b','c','m','l','k','j','i','h','g',])
y1=np.array([4,1,9,4,5,2,7,1,6,8]) 
plt.bar(x1,y1)
```

```python
#horizontl
x1=np.array(['a','b','c','m','l','k','j','i','h','g',])
y1=np.array([4,1,9,4,5,2,7,1,6,8]) 
plt.barh(x1,y1)
```

```python
#color
x1=np.array(['a','b','c','m','l','k','j','i','h','g',])
y1=np.array([4,1,9,4,5,2,7,1,6,8]) 
plt.barh(x1,y1,color='purple')
```

```python
#width
x1=np.array(['a','b','c','m','l','k','j','i','h','g',])
y1=np.array([4,1,9,4,5,2,7,1,6,8]) 
plt.bar(x1,y1,color='purple',width= 0.2)
```

## Piechart

```python

y1=np.array([4,1,9,4,5,2,7,1,6,8]) 
plt.pie(y1)
```

```python
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
plt.pie(y1,labels=mylabel) # labeling
```

```python
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
plt.pie(y1,labels=mylabel,startangle=90) #start angle
```

```python
#exploding a part 
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
myexplode=(0.2,0.3,0.1,0,0)
plt.pie(y1,labels=mylabel,startangle=90,explode=myexplode)
```

```python
#3d
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
myexplode=(0.2,0.3,0.1,0,0)
plt.pie(y1,labels=mylabel,startangle=90,explode=myexplode,shadow=True)
```

```python
#color change#3d
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
color=(['violet','indigo','blue','green','yellow'])
myexplode=(0.2,0.3,0.1,0,0)
plt.pie(y1,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=color)
```

```python
#autopacity=value
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
color=(['violet','indigo','blue','green','yellow'])
myexplode=(0.2,0.3,0.1,0,0)
plt.pie(y1,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=color,autopct='%1.1f%%')
```

```python
#legend : display color block
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
color=(['violet','indigo','blue','green','yellow'])
myexplode=(0.2,0.3,0.1,0,0)
plt.pie(y1,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=color,autopct='%1.1f%%')
plt.legend()
```

```python
#titled
y1=np.array([4,1,9,4,5])
mylabel=('grapes','mango','orange','apple','pineapple') 
color=(['violet','indigo','blue','green','yellow'])
myexplode=(0.2,0.3,0.1,0,0)
plt.pie(y1,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=color,autopct='%1.1f%%')
plt.legend(title='color')
```
