```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("copy.csv")
df.tail()
```

```python
# Count number of employees in each department
dept_counts = df['Department\nHR'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(dept_counts.index, dept_counts.values, color=['blue', 'purple', 'orange'])
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees in Each Department")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding grid for better visibility
plt.show()
```

```python
# Filter employees with a salary greater than 60,000
high_salary_employees = df[df["Salary\n50000"] > 60000]

# Display filtered data
high_salary_employees
```

```python
# Group by Department and calculate average salary
dept_salary = df.groupby("Department\nHR")["Salary\n50000"].mean().reset_index()

# Plot the line graph
plt.figure(figsize=(8, 5))
plt.plot(dept_salary["Department\nHR"], dept_salary["Salary\n50000"], marker='o', linestyle='-', color='red')
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Salary Trend by Department")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adding grid for better visibility
plt.show()
```
