 # Relay plot

```python
sns.relplot(x='total_bill',y='tip',data=data) #scatter plot of tip nd bill
```

```python
data['time'].unique() #unique value
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',data=data)    #viewing the no of customer visiting in lunch and dinner
# used to display comaprison
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',data=data)  # used to differntiate btw smoker or non smoker 
```

```python
# differnt graph of smoker
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',kind='scatter',col='smoker',data=data)  # kind for multiple graph , col for the reason
```

```python
# differnt graph of time
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',kind='scatter',col='time',data=data) 
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',kind='scatter',row='time',data=data) #row for row representaion
```

```python
data['day'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='time',style='smoker',kind='scatter',col='day',data=data) # for multiple graph acc tO features unique value
```

```python
# xlabel ,ylabel,tile
sns.relplot(x='total_bill',y='tip',data=data)
plt.title('total_bill vs tip')
plt.xlabel("total_bill",fontsize=20,color='purple')
plt.ylabel('tip',fontsize=20,color='indigo')
```

# Cat plot

```python
sns.catplot(x='day',y='total_bill',data=data,kind='swarm') # display catplot and swarn to disply horizontal data
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='strip') # strip to display dataset
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='box')   #represnt the maximum range of the day
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='bar')  # display the max range
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='violin')  # display the max range
```

```python
sns.catplot(x='day',y='total_bill',data=data,kind='boxen') # display the max range
```
